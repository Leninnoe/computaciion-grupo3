```python
import numpy as np
import matplotlib.pyplot as plt

# Definir las funciones
def f1(x):
    return np.sin(x)

def f2(x):
    return np.exp(3 * x)

def f3(x):
    return 1 / x

# Definir los rangos
x1 = np.linspace(-2, 2, 400)
x2 = np.linspace(-2, 4, 400)
x3 = np.linspace(1, 10, 400)

# Graficar las funciones
plt.figure(figsize=(12, 8))

# Primera funcin
plt.subplot(3, 1, 1)
plt.plot(x1, f1(x1), label='sin(x)')
plt.title('f(x) = sin(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()

# Segunda funcin
plt.subplot(3, 1, 2)
plt.plot(x2, f2(x2), label='e^(3x)', color='red')
plt.title('f(x) = e^(3x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()

# Tercera funcin
plt.subplot(3, 1, 3)
plt.plot(x3, f3(x3), label='1/x', color='green')
plt.title('f(x) = 1/x')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()

# Ajustar el espacio entre los subplots
plt.tight_layout()

# Mostrar las grficas
plt.show()
```
