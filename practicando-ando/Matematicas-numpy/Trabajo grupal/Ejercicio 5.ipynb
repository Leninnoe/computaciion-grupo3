```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import lagrange

# Puntos de datos
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([12, 14, 22, 39, 58, 77])

# Crear el polinomio de Lagrange
polinomio = lagrange(x, y)

# Mostrar el polinomio
print("Polinomio de Lagrange:")
print(polinomio)

# Evaluar el polinomio en puntos nuevos
x_nuevo = np.linspace(0, 5, 100)
y_nuevo = polinomio(x_nuevo)

# Graficar los puntos de datos y el polinomio de Lagrange
plt.scatter(x, y, color='red', label='Datos originales')
plt.plot(x_nuevo, y_nuevo, label='Polinomio de Lagrange')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('Interpolación con Polinomio de Lagrange')
plt.grid(True)
plt.show()
```
