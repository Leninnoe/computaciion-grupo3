```python
import numpy as np
import matplotlib.pyplot as plt

# Definir el dominio
x = np.linspace(-1, 3, 400)

# Definir la función
f_x = np.exp(-x**2)

# Crear la gráfica
plt.plot(x, f_x, label='Función f(x)')

# Añadir título y etiquetas
plt.title('Función f(x)')
plt.xlabel('Eje x')
plt.ylabel('f(x)')

# Añadir leyenda
plt.legend()

# Mostrar la gráfica
plt.show()
```
