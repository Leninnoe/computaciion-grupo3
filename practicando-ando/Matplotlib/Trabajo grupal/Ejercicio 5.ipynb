```python
import numpy as np
import matplotlib.pyplot as plt

def graficar_senales(f1=10, f2=100):
    
    t = np.linspace(0, 0.5, 1000)
    
   
    señal = np.sin(2 * np.pi * f1 * t) + np.sin(2 * np.pi * f2 * t)
    
    
    plt.figure(figsize=(10, 5))
    plt.plot(t, señal, label='Señal', color='orangered')
    plt.title('Dos frecuencias')
    plt.xlabel('Tiempo (t)')
    plt.legend()
    plt.grid(True)
    plt.style.use('ggplot')  
    plt.show()


graficar_senales()
```
