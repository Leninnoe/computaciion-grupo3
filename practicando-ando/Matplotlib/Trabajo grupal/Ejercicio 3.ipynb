```python
import matplotlib.pyplot as plt
import numpy as np

# Generar datos
x = np.random.rand(100)
y = np.random.rand(100)
sizes = np.random.randint(10, 200, size=100)  # Tamaos aleatorios para los puntos
colors = np.random.rand(100)  # Colores aleatorios para los puntos

# Graficar la nube de puntos
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.scatter(x, y, s=sizes, c=colors, cmap='viridis')
plt.colorbar(label='Valores')
plt.title('Grfica con colormap viridis')

plt.subplot(1, 2, 2)
plt.scatter(x, y, s=sizes, c=colors, cmap='plasma')
plt.colorbar(label='Valores')
plt.title('Grfica con colormap plasma')

plt.tight_layout()
plt.show()
```
